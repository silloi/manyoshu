```python
import requests
import csv
import re
from bs4 import BeautifulSoup
import codecs
import ruamel_yaml
import time

yaml = ruamel_yaml.YAML()

# スクレイピング対象の URL にリクエストを送り HTML を取得する
res = requests.get('https://ja.wikisource.org/wiki/%E4%B8%87%E8%91%89%E9%9B%86/%E7%AC%AC%E4%BA%8C%E5%8D%81%E5%B7%BB')

# レスポンスの HTML から BeautifulSoup オブジェクトを作る
soup = BeautifulSoup(res.text, 'html.parser')

# span 要素を全て取得する
titles = soup.find_all('span', {'id': re.compile('[0-9]{2}/[0-9]{4}')})

for data in titles:
  if (len(data.parent.find_next().find_next().text.split('/')) > 1) and not (data.parent.find_next().find_next().text.split('/')[1].isdigit()):
    continue
    
  id = data.parent.find_next().find_next().text.split('/')[1] if len(data.parent.find_next().find_next().text.split('/')) > 1 else data.parent.find_next().find_next().text.split('/')[0]
    
  dict = {"title": "", "description": "", "position": 0, "category": "", "version": "", "volume": 0, "comparison": [], "keywords": []}
  dict["position"] = int(data.parent.find_next().find_next().text.split('/')[1]) if len(data.parent.find_next().find_next().text.split('/')) > 1 else 0
  dict["version"] = "1.0"
  dict["volume"] = int(data.parent.find_next().find_next().text.split('/')[0]) if len(data.parent.find_next().find_next().text.split('/')) > 1 else 0
  dict["category"] = "巻" + str(dict["volume"])
  dict["title"] = data.parent.find_next().find_next().find_next().text.strip("[題詞]").strip("""

  """)
  original = data.parent.find_next().find_next().find_next().find_next().text.strip("[原文]").strip("""

  """)
  reading = data.parent.find_next().find_next().find_next().find_next().find_next().text.strip("[訓読]").strip("""

  """)
  kana = data.parent.find_next().find_next().find_next().find_next().find_next().find_next().text.strip("[仮名]").strip("""

  """)
  note = data.parent.find_next().find_next().find_next().find_next().find_next().find_next().find_next().text.strip("[左注]").strip("""

  """)
  dict["comparison"] = data.parent.find_next().find_next().find_next().find_next().find_next().find_next().find_next().find_next().text.strip("[校異]").strip("""

  """).split(' / ')
  dict["keywords"] = data.parent.find_next().find_next().find_next().find_next().find_next().find_next().find_next().find_next().find_next().text.strip("[事項]").strip("""

  """).split(' ')
  
  dict["description"] = reading

  # list.append(dict)

  print(dict["title"])

  with codecs.open(id + '.md', 'w', 'utf-8') as f:
    f.write('---\n')
    yaml.dump(dict, f)
    time.sleep(0.1)
    f.write('---\n')
    f.write("""
## 原文

{}
""".format(original))
    f.write("""
## 訓読

{}
""".format(reading))
    f.write("""
## 仮名

{}
""".format(kana))
    f.write("""
## 左注

{}
""".format(note))
    f.close()
```
